```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the CSV file
data = pd.read_csv('sales_data.csv')

# Kiểm tra các cột trong dữ liệu
print(data.columns)

# Feature Scaling
standard_scaler = StandardScaler()

# Chỉ định các cột số cần chuẩn hóa
num_cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Longitude', 'Latitude']
data[num_cols] = standard_scaler.fit_transform(data[num_cols])

# Label Encoding cho các đặc trưng phân loại
le_item_fat_content = LabelEncoder()
data['Item_Fat_Content'] = le_item_fat_content.fit_transform(data['Item_Fat_Content'])

le_item_type = LabelEncoder()
data['Item_Type'] = le_item_type.fit_transform(data['Item_Type'])

le_outlet_identifier = LabelEncoder()
data['Outlet_Identifier'] = le_outlet_identifier.fit_transform(data['Outlet_Identifier'])

le_outlet_size = LabelEncoder()
data['Outlet_Size'] = le_outlet_size.fit_transform(data['Outlet_Size'])

le_outlet_location_type = LabelEncoder()
data['Outlet_Location_Type'] = le_outlet_location_type.fit_transform(data['Outlet_Location_Type'])

le_outlet_type = LabelEncoder()
data['Outlet_Type'] = le_outlet_type.fit_transform(data['Outlet_Type'])

# Label Encoding cho Item_Outlet_Sales
# Nếu bạn vẫn muốn mã hóa Item_Outlet_Sales, hãy chú ý rằng đây không phải là cách thường dùng cho biến số liên tục.
le_item_identifier = LabelEncoder()
# Mã hóa cột Item_Identifier
data['Item_Identifier'] = le_item_identifier.fit_transform(data['Item_Identifier'])
# Drop rows with missing values
data.dropna(inplace=True)

def plotCorrelationMatrix(data):
    # Tính toán ma trận tương quan
    corr = data.corr()
    
    # Vẽ heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Ma trận tương quan")
    plt.show()

# Sử dụng hàm để vẽ ma trận tương quan
plotCorrelationMatrix(data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
import pickle
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv('sales_data.csv')

# Initialize LabelEncoders for all necessary categorical features
le_item_identifier = LabelEncoder()
le_outlet_identifier = LabelEncoder()
le_outlet_size = LabelEncoder()
le_outlet_location_type = LabelEncoder()
le_outlet_type = LabelEncoder()

# Fit the LabelEncoders on the respective columns
data['Item_Identifier'] = le_item_identifier.fit_transform(data['Item_Identifier'])
data['Outlet_Identifier'] = le_outlet_identifier.fit_transform(data['Outlet_Identifier'])
data['Outlet_Size'] = le_outlet_size.fit_transform(data['Outlet_Size'])
data['Outlet_Location_Type'] = le_outlet_location_type.fit_transform(data['Outlet_Location_Type'])
data['Outlet_Type'] = le_outlet_type.fit_transform(data['Outlet_Type'])

# Select features and target
X = data[['Item_Identifier', 'Outlet_Identifier', 'Outlet_Size', 'Item_MRP', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Visibility']]
y = data['Item_Outlet_Sales']

# Split data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train Linear Regression model
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("Linear Regression Root Mean Squared Error: {:,.2f}".format(error))

# Train RandomForest model
regressor = RandomForestRegressor(random_state=0)
parameters = {
    "max_depth": [None, 2, 4, 6, 8, 10, 12],
    "n_estimators": [10, 50, 100]  # Added hyperparameter for tuning
}
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error', cv=5)  # Added cross-validation
gs.fit(x_train, y_train)
best_estimator = gs.best_estimator_

# Evaluate model
y_pred = best_estimator.predict(x_test)
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("RandomForest Root Mean Squared Error: {:,.2f}".format(error))

# Save model and encoders
data_to_save = {
    "model": best_estimator,
    "le_item_identifier": le_item_identifier,
    "le_outlet_size": le_outlet_size,
    "le_outlet_identifier": le_outlet_identifier,
    "le_outlet_location_type": le_outlet_location_type,
    "le_outlet_type": le_outlet_type
}

with open('saved_steps.pkl', 'wb') as file:
    pickle.dump(data_to_save, file)
```

```python
item_mrp_values = np.arange(0, 300, 100)  # Thay đổi khoảng giá trị tùy theo nhu cầu
predicted_sales = []

for mrp in item_mrp_values:
    # Tạo mẫu dữ liệu mới với Item_MRP thay đổi
    columns = ["Item_Identifier", "Outlet_Identifier", "Outlet_Size","Item_MRP", "Outlet_Location_Type", "Outlet_Type", "Item_Visibility"]
    x_new = pd.DataFrame([["FDU02","OUT035","High", mrp, "Tier 2", "Supermarket Type1", -1.092117e-16]], columns=columns)
    
    # Chuyển đổi dữ liệu cho phù hợp với biến đầu vào của mô hình
    x_new["Item_Identifier"] = le_item_identifier.transform(x_new["Item_Identifier"])
    x_new["Outlet_Size"] = le_outlet_size.transform(x_new["Outlet_Size"])
    x_new["Outlet_Identifier"] = le_outlet_identifier.transform(x_new["Outlet_Identifier"])
    x_new["Outlet_Location_Type"] = le_outlet_location_type.transform(x_new["Outlet_Location_Type"])
    x_new["Outlet_Type"] = le_outlet_type.transform(x_new["Outlet_Type"])
    
    # Dự đoán Item_Outlet_Sales cho mẫu dữ liệu mới
    y_pred_new = best_estimator.predict(x_new)
    predicted_sales.append(y_pred_new[0])

# Vẽ đồ thị Partial Dependence Plot
plt.plot(item_mrp_values, predicted_sales, marker='o')
plt.xlabel('Item_MRP')
plt.ylabel('Predicted Item_Outlet_Sales')
plt.title('Partial Dependence Plot for Item_MRP (RandomForest)')
plt.grid()
plt.show()
```
